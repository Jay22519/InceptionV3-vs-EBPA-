```python
import tensorflow as tf 
```

```python
from tensorflow.keras.preprocessing import image
```

```python
import numpy as np
```

```python
model = tf.keras.applications.InceptionV3(
    include_top=True,
    weights="imagenet",
    input_tensor=None,
    input_shape=None,
    pooling=None,
    classes=1000
)
```

```python
img_path = 'Elephant.PNG'
```

```python
img = image.load_img(img_path, target_size=(299, 299,3))
```

```python
img
```

```python
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = x = tf.keras.applications.mobilenet.preprocess_input(x)
```

```python
preds = model.predict(x)
```

```python
tf.keras.applications.inception_v3.decode_predictions(
    preds, top=5
)
```
